# Дополнительные возможности

Мы уже упоминали, что булевы значения True и False можно рассматривать как единицы и нули. Мы можем использовать это свойство в создании условий. То есть значение самой переменной будет восприниматься как False, если она равна нулю, и как True во всех других случаях. 

Рассмотрим пример с барменом. Продажа напитка зависит не только от возраста, но и от того, есть ли у клиента деньги. Напишем программу, которая сразу и без обиняков выясняет у клиента его платежеспособность. Если есть хоть сколько-то денег, мы предлагаем ему совершить покупку, есть денег нет – отказываем.
Вместо сравнения переменной money с каким-то значением, мы используем её саму. Нас устроит любой вариант кроме нуля. 


```python
money = int(input("Сколько у вас денег"))
if money:
    print("Мы рады предложить вам большой выбор напитков")
else:
    print("У вас недостаточно денег")
```

Python понял значение 55 как True и выбрал первый путь. Теперь проверим с нулём:


```python
money = int(input("Сколько у вас денег"))
if money:
    print("Мы рады предложить вам большой выбор напитков")
else:
    print("У вас недостаточно денег")
```

Во всех примерах, которые мы разбирали, единственная наша инструкция была напечатать что-нибудь на экране. Мы это делали только чтобы понять, по какому из путей пойдет Python. Но мы совершенно не ограничены в том, какие инструкции ставить на каждом из путей. Дополним наш последний пример другими инструкциями. 

Например, если у человека есть деньги, мы будем вычитать из них какую-то сумму (стоимость напитка) и выводить оставшееся на экран. Для этого перезапишем переменную, вычитая сумму из её предыдущего значения:



```python
money = int(input("Сколько у вас денег"))
if money:
    print("Мы рады предложить вам большой выбор напитков")
    print("Все напитки по 15")
    money = money - 15
    print("У вас осталось", money, "денег")
else:
    print("У вас недостаточно денег")
```

После любого из операторов if, elif, else может быть инструкция любой длины и сложности.

Бывают случаи, когда нужно задать какую-то инструкцию при выполнении условия, но для противного случая у вас указаний нет, не требуется делать ничего. В таком случае вы можете воспользоваться оператором if без оператора else. 

Дополним наш предыдущий пример ещё одним предложением напитка в случае, если у клиента осталось достаточно денег. Добавляем конструкцию с if и проверкой количества денег. Если денег недостаточно, никаких инструкций не даём.

```python
money = int(input("Сколько у вас денег"))
if money:
    print("Мы рады предложить вам большой выбор напитков")
    print("Все напитки по 15")
    money = money - 15
    print("У вас осталось", money, "денег")
    if money >= 15:
        print("Хотите повторить?")
else:
    print("У вас недостаточно денег")
```

Проверим для меньшего количества денег:


```python
money = int(input("Сколько у вас денег"))
if money:
    print("Мы рады предложить вам большой выбор напитков")
    print("Все напитки по 15")
    money = money - 15
    print("У вас осталось", money, "денег")
    if money >= 15:
        print("Хотите повторить?")
else:
    print("У вас недостаточно денег")
```

Последняя выполненная инструкция была вывести баланс. А поскольку следующее условие не было выполнено, то и сделано ничего не было.
