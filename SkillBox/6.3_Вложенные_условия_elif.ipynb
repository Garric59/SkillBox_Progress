# Вложенные условия, elif

Вернёмся к примеру про животных в самолёте. Мы поделили их на тех, кто полетит в салоне, и тех, кто будет сидеть в багажном отсеке. Но и в салоне самолёта можно передвигаться по-разному: чьи-то переноски ставят под сиденье пассажира, а другие закрепляют на отдельном кресле (у некоторых авиакомпаний). Это зависит от размера животного. Давайте добавим это условие.

Нас интересуют только животные в салоне, поэтому мы будем изменять только инструкции, относящиеся к этому случаю. А также добавим переменную для значения веса:

```python
animal = "кошка"
weight = 10
if animal == "собака" or animal == "кошка":
    print("Вы можете взять своего питомца в салон")
    if weight <= 5:
        print("Ваш питомец будет размещен под креслом")
    else:
        print("Мы закрепим переноску на отдельном кресле")
else:
    print("Извините, ваш питомец полетит в багажном отсеке")
```

Первое условие выполнено, второе не выполнено. Поэтому в первом случае программа пошла по первому пути, а на второй развилке свернула на второй путь.

Проверим с меньшим весом:


```python
animal = "кошка"
weight = 3
if animal == "собака" or animal == "кошка":
    print("Вы можете взять своего питомца в салон")
    if weight <= 5:
        print("Ваш питомец будет размещен под креслом")
    else:
        print("Мы закрепим переноску на отдельном кресле")
else:
    print("Извините, ваш питомец полетит в багажном отсеке")
```

Оба условия выполнены, и программа оба раза выбирает первый путь.

Проверим с другим животным:


```python
animal = "хорек"
weight = 3
if animal == "собака" or animal == "кошка":
    print("Вы можете взять своего питомца в салон")
    if weight <= 5:
        print("Ваш питомец будет размещен под креслом")
    else:
        print("Мы закрепим переноску на отдельном кресле")
else:
    print("Извините, ваш питомец полетит в багажном отсеке")
```

Первое условие не было выполнено, поэтому программа даже не стала проверять, что там с весом, и сразу же ушла в инструкции для else.

В случае вложенных условий мы проходим несколько последовательных развилок, каждый раз выбирая один из двух путей.
Есть и другой вариант усложнения, когда развилка содержит больше двух путей. Для этого используется оператор `elif`. 

Напишем программу, которая будет с нами здороваться в соответствии со временем суток.
Всё, что до 5 утра, будем считать ночью. Всё, что до полудня – утром. До 18 включительно у нас будет день, а всё остальное – вечер.

Переменную запросим с помощью input().



```python
time = int(input("Который час?"))
if time < 5:
    print("Доброй ночи")
elif time < 12:
    print("Доброе утро")
elif time <= 18:
    print("Добрый день")
else:
    print("Добрый вечер")
```

```python
time = int(input("Который час?"))
if time < 5:
    print("Доброй ночи")
elif time < 12:
    print("Доброе утро")
elif time <= 18:
    print("Добрый день")
else:
    print("Добрый вечер")
```

Python поочерёдно сравнивает все указанные условия и идёт по подходящему пути. Хотя 9 подходит и под второе, и под третье условие, выбирается второй путь, поскольку именно он идёт вначале. Третий путь рассматривает только те случаи, которые уже не подошли под первый и второй. То есть туда попадали бы только те числа, которые больше или равны 12 и меньше или равны 18.
