В поле URL есть год выпуска проверим совпадения с полем Year

```python
# В поле URL содержится год выпуска авто. Вытащим его в новую переменную YEAR_PARSE.
p1 = r"(?:[A-z\./]+-)([\d]{4})"


def match_year(text, pattern=p1):
    matches = re.search(pattern, text)
    if matches:
        return int(matches.group(1))


df_work['year_parse'] = df_work.url.apply(match_year)
# Проверим заполнение
df_work[df_work.year_parse.isna()][['url', 'year_parse', 'year']]
```

```python
# Проверка совпадений новой переменной year_parse с существующей year
print('Значения совпадают на {}%'.format(
    round(df_work[df_work.year == df_work.year_parse]['year'].count() /
          df_work.shape[0] * 100)))
```

```python
# Объединим оба столбца и запишем в новый year_full
def year_full(original_year, parse_year,
              category):  # Функция определения года выпуска авто
    if original_year > 0.0:
        return original_year
    else:
        if parse_year > 0.0:
            return parse_year
        else:
            # Возвращает самое часто встречающееся значение в нужной категории
            return df_work[df_work['price_category'] ==
                           category]['year'].mode().values[0]
```

```python
df_work[
    'year_full'] = df_work.loc[:,
                               ['year', 'year_parse', 'price_category']].apply(
                                   lambda row: year_full(
                                       row['year'], row['year_parse'], row[
                                           'price_category']),
                                   axis=1)
# Проверим заполнение
df_work[(df_work.year.isna()) & (df_work.year_parse.isna())][[
    'url', 'year_parse', 'year_full', 'year', 'price_category'
]]
```

```python
# Проверим результат заполнения значений
print('Количество не заполненых значений: {}'.format(
    df_work.year_full.isna().sum()))
```

#### Обработка пропусков в типе топлива автомобиля

```python
# Количество значений для типа топлива, включая пропущенные.
df_work.fuel.value_counts(dropna=False)
```

```python
# Результат запишем в новый столбец fuel_full
def full_fuel(fuel_, manufacturer, category):
    # Функция определения топлива авто по категории и производителю.
    # Функция заполняет пропуски самым популярным значением, так же обработаем Other.
    if type(fuel_) != float and fuel_ != 'other':
        return fuel_
    elif fuel_ == 'other' or type(fuel_) == float:
        # Возвращает самое часто встречающееся значение в нужной категории
        return df_work[(df_work.price_category == category) & (
            df_work.manufacturer == manufacturer)]['fuel'].mode().values[0]
```

```python
df_work['fuel_full'] = df_work.loc[:,
                                   ['fuel', 'manufacturer', 'price_category'
                                    ]].apply(lambda row: full_fuel(
                                        row['fuel'], row['manufacturer'], row[
                                            'price_category']),
                                             axis=1)
# Проверим заполнение
df_work[(df_work.fuel.isna()) & (df_work.fuel_full.isna())][[
    'fuel', 'fuel_full', 'price_category', 'manufacturer'
]]
```

```python
# Посмотрим на распределение переменных fuel_full, fuel после замены
fig4, (fuel_full_bar, fuel) = plt.subplots(1, 2, figsize=(36, 12))
df_work.fuel_full.hist(bins=10, color='red', ax=fuel_full_bar)
df_work.fuel.hist(bins=10, color='green', ax=fuel)
fuel_full_bar.set_ylabel('Years')
fuel_full_bar.set_xlabel('count')
fuel_full_bar.set_title('Count fuel_full')
fuel.set_title('Count fuel')
fuel.set_xlabel('count')
```

```python
# Посчитаем распределение новой переменной относительно целевой
stats = df_work.groupby(['fuel_full', 'price_category'],
                        as_index=False)[['id']].count()
stats
```

```python
# Визуализируем распределение переменной fuel_full после замены относительно price_category
fig5, (fuel_full_bar, fuel_bar) = plt.subplots(1, 2, figsize=(36, 12))
stats_ff = df_work.groupby(['fuel_full', 'price_category'],
                           as_index=False)[['id']].count()
stats_ff = stats_ff.rename(columns={'id': 'count'})
stats_f = df_work.groupby(['fuel', 'price_category'],
                          as_index=False)[['id']].count()
stats_f = stats_f.rename(columns={'id': 'count2'})
sns.barplot(y='count',
            x='price_category',
            hue='fuel_full',
            data=stats_ff,
            ax=fuel_full_bar)
sns.barplot(y='count2',
            x='price_category',
            hue='fuel',
            data=stats_f,
            ax=fuel_bar)
fuel_full_bar.set_ylabel('Count')
fuel_full_bar.set_xlabel('price_category')
fuel_full_bar.set_title('Count fuel_full for price_category')
fuel_bar.set_xlabel('price_category')
fuel_bar.set_title('Count fuel for price_category')
```

```python
# Функция поиска значения в справочнике и при ошибке в написании.
def check_value(value, values_reference):
    for item in values_reference:
        r = SequenceMatcher(None, value, item).ratio()
        if item == value:
            return item
        elif r > .7:
            return item
        else:
            continue
    return '?'
```

```python
# Функция прохода по списку и проверка значения в справочнике.
def check_list_one_item(list_value, values_reference):
    for x in list_value:
        if x == 'vw':
            x = 'volkswagen'
        elif x == 'chevy':
            x = 'chevrolet'
        elif x == 'ram':
            x = 'dodge'
        elif x == 'silverado':
            x = 'chevrolet'
        if check_value(x, values_reference) != '?':
            return check_value(x, values_reference)
    return '?'
```

```python
# Функция прохода по списку и проверка значения в справочнике.
def check_list_two_item(list_value, values_reference):
    if len(list_value) > 1:
        for x, y in zip(list_value[0::1], list_value[1::1]):
            if x == 'vw':
                x = 'volkswagen'
            elif x == 'chevy':
                x = 'chevrolet'
            elif x == 'ram':
                x = 'dodge'
            elif x == 'silverado':
                x = 'chevrolet'
            if y == 'vw':
                y = 'volkswagen'
            elif y == 'chevy':
                y = 'chevrolet'
            elif y == 'ram':
                y = 'dodge'
            elif y == 'silverado':
                y = 'chevrolet'
            if check_value(x + '-' + y, values_reference) != '?':
                return check_value(x + '-' + y, values_reference)
    return '?'
```

```python
# Соберем производителя авто в новую переменную manufacturer_full
def fill_manufacturer(id_, manufacturer, url, region_url, model, df_=df_m):
    model_parse = model.lower().replace(',', '').split()  # Парсим model
    # Парсим URL
    url_parse = [
        num for elem in [
            i.split('-') for i in url.replace(region_url, '').split('/')
            if i.__contains__('-')
        ] for num in elem
    ]
    if manufacturer == 'other':
        values_reference = df_.name.unique()
    else:
        values_reference = df_m.loc[df_m['name'].str[0] ==
                                    manufacturer[0]].name.unique()
    try:
        # Проверка на наличие значения переменной manufacturer в справочнике
        if check_value(manufacturer, values_reference) != '?':
            return check_value(manufacturer, values_reference)
        # Проверка на наличие значения элемента списка URL_PARSE в справочнике
        elif check_list_one_item(url_parse, values_reference) != '?':
            return check_list_one_item(url_parse, values_reference)
        # Проверка на наличие значения элементов списка URL_PARSE в справочнике
        elif check_list_two_item(url_parse, values_reference) != '?':
            return check_list_two_item(url_parse, values_reference)
        # Проверка на наличие значения элемента списка MODEL_PARSE в справочнике
        elif check_list_one_item(model_parse, values_reference) != '?':
            return check_list_one_item(model_parse, values_reference)
        # Проверка на наличие значения элементов списка MODEL_PARSE в справочнике
        elif check_list_two_item(model_parse, values_reference) != '?':
            return check_list_two_item(model_parse, values_reference)
        else:
            # Не нашли соответствий
            return 'other'

    except Exception as err:
        print("На этой строке ошибка: {} url_parse:{} model_parse:{} err:{}".
              format(id_, url_parse, model_parse, err))
    return '?'
```

```python
# Соберем производителя авто в новую переменную manufacturer_full
def fill_model(id_, manufacturer, url, region_url, model, df_=df_m):
    # Парсим model
    model_parse = model.lower().replace(',', '').split()  # Парсим model
    # Парсим URL
    url_parse = [
        num for elem in [
            i.split('-') for i in url.replace(region_url, '').split('/')
            if i.__contains__('-')
        ] for num in elem
    ]
    try:
        if len(df_m[df_m.name == manufacturer].models) > 0:
            model_unique = pd.DataFrame(
                df_[df_.name == manufacturer].models.values[0]).name.apply(
                    lambda x: x.lower())
        else:
            return 'other'
        # if manufacturer != 'other':
        #     model_unique = df_[df_.make == manufacturer].model.unique()
        # else:
        #     model_unique = df_.model.unique()
        # Проверка на наличие значения переменной manufacturer в справочнике
        if check_value(model, model_unique) != '?':
            return check_value(model, model_unique)
        # Проверка на наличие значения элемента списка MODEL_PARSE в справочнике
        elif check_list_one_item(model_parse, model_unique) != '?':
            return check_list_one_item(model_parse, model_unique)
        # Проверка на наличие значения элементов списка MODEL_PARSE в справочнике
        elif check_list_two_item(model_parse, model_unique) != '?':
            return check_list_two_item(model_parse, model_unique)
        # Проверка на наличие значения элемента списка URL_PARSE в справочнике
        elif check_list_one_item(url_parse, model_unique) != '?':
            return check_list_one_item(url_parse, model_unique)
        # Проверка на наличие значения элементов списка URL_PARSE в справочнике
        elif check_list_two_item(url_parse, model_unique) != '?':
            return check_list_two_item(url_parse, model_unique)
        else:
            # Не нашли соответствий
            return 'other'
    except Exception as err:
        print("На этой строке ошибка: {} url_parse:{} model_parse:{} err:{}".
              format(id_, url_parse, model_parse, err))
```

```python
# Тест Граббса
# Автор реализации теста Граббса и теста Рознера для Python
# Опишем три функции: первая находит значение критерия Граббса и максимальное значение в наборе данных, вторая —
# критическое значение с учётом объёма выборки и уровня значимости, а третья проверяет, является ли значение с максимальным индексом выбросом:
from scipy import stats


def grubbs_stat(y):
    std_dev = np.std(y)
    avg_y = np.mean(y)
    abs_val_minus_avg = abs(y - avg_y)
    max_of_deviations = max(abs_val_minus_avg)
    min_of_deviations = min(abs_val_minus_avg)
    max_ind = np.argmax(abs_val_minus_avg)
    min_ind = np.argmin(abs_val_minus_avg)
    Gcal_max = max_of_deviations / std_dev
    Gcal_min = min_of_deviations / std_dev
    print(f"Grubbs Statistics Value: {Gcal_max, Gcal_min}")
    return Gcal_max, max_ind, Gcal_min, min_ind


def calculate_critical_value(size, alpha):
    t_dist = stats.t.ppf(1 - alpha / (2 * size), size - 2)
    numerator = (size - 1) * np.sqrt(np.square(t_dist))
    denominator = np.sqrt(size) * np.sqrt(size - 2 + np.square(t_dist))
    critical_value = numerator / denominator
    print(f"Grubbs Critical Value: {critical_value}")
    return critical_value


def check_G_values_max(Gs, Gc, inp, max_index):
    if Gs > Gc:
        return inp[max_index]
    else:
        print(f"{inp[max_index]} is not an outlier")


def check_G_values_min(Gs, Gc, inp, min_index):
    print(Gs, Gc)
    if Gs < Gc:
        return inp[min_index]
    else:
        print(f"{inp[min_index]} is not an outlier")
```

```python
# Определим границы с помощью функции.
def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries_outliers = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)

    return boundaries_outliers
```

```python
# Определим границы с помощью функции.
def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries_outliers = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)

    return boundaries_outliers
```
